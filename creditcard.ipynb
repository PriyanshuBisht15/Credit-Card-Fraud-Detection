# 9. Handling Imbalanced Dataset

```python
# 1.Undersampling
# 2.Oversampling
```

# Undersampling

```python
normal=data[data['Class']==0]
fraud=data[data['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=492)
```

```python
normal_sample.shape
```

```python
new_data=pd.concat([normal_sample,fraud],ignore_index=True)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
x=new_data.drop('Class',axis=1)
y=new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

# 10. Logistic Regression

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Assuming you have your data ready in variables X and y
# X is your feature matrix, and y is your target vector



# Scale the features (optional, but recommended for Logistic Regression)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Create a LogisticRegression model
log = LogisticRegression()

# Fit the model to the training data
log.fit(x_train, y_train)

# Predict using the test data
y_pred1 = log.predict(x_test)

# Evaluate accuracy (or use other metrics like precision, recall, etc.)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred1)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```
